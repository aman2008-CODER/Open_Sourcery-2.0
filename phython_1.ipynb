```python
# wap to reverse a number
n=int(input("enter the value "))
a=n
r=0
while(n>0):
  ld=n%10
  r=r*10+ld
  n=n//10
print("the reverse of a number is",r)

# Removed the line 'r=b' as 'b' was undefined and 'r' already holds the reversed number.
# Changed 'if(a==b)' to 'if(a==r)' to compare the original number with its reverse.
if(a==r):
  print("the number is palindrome")
else:
  print("the number is not palindrome")
```
